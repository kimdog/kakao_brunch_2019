```python
# ------------------------------------------------------------------------------
# get word dictionary 
# type 1 : { word : count }
# type 2 : { word : doc_id }
# ------------------------------------------------------------------------------

import os 
TEMP_FOLDER = 'tmp/'
RES_FOLDER = 'res/'
from smart_open import open
import json

SIZE_OF_CHUNK = 5000

# contain all word, contain all docid
def read_corpus(f, ss1, ss2): 
    count = 0;
    for jsonLine in f :
        
        line = json.JSONDecoder().decode(jsonLine)
        doc_id = line['id']
        
        if count == 0 : print(doc_id)
        
        for morphs in line['morphs'] :
            #yield gensim.models.doc2vec.TaggedDocument(morphs, [doc_id])  
            ss2[doc_id] = morphs[0]
            for mor in morphs :
                #ss[mor] = ss.get(mor, 0) + 1
                ss1[mor] = doc_id
        
        count = count + 1
        if count >= SIZE_OF_CHUNK : break
            
#file open_learning data file 
resFolder = 'res/'
contentsSize = 7
contents = [ open(os.path.join(RES_FOLDER, 'contents/data.' + str(i)), 'r') for i in range(contentsSize) ]
contentsLineSize = [ 100000, 100000, 100000, 100000, 100000, 100000, 42190 ]

tmpDic1 = {}
tmpDic2 = {}

for i in range(contentsSize) :    
    for j in range( int(( contentsLineSize[i] + SIZE_OF_CHUNK - 1 ) / SIZE_OF_CHUNK ) ) :
        # read corpus
        read_corpus(contents[i], tmpDic1, tmpDic2)
        
            
ff = open(TEMP_FOLDER+'wordDic_all_word', 'w')

ff.write(json.JSONEncoder().encode(tmpDic1))

ff.close()

ff = open(TEMP_FOLDER+'wordDic_all_docid', 'w')

ff.write(json.JSONEncoder().encode(tmpDic2))

ff.close()


ff = open(TEMP_FOLDER+'wordDic_all_word', 'r')

mydic1 = json.JSONDecoder().decode(ff.read())

ff.close()

ff = open(TEMP_FOLDER+'wordDic_all_docid', 'r')

mydic2 = json.JSONDecoder().decode(ff.read())

ff.close()


```

```python
# ------------------------------------------------------------------------------
# get doc2vec
# from contents/read.0 ~ 6 files
# ------------------------------------------------------------------------------

import logging 
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO) 
import os 
import tempfile 
TEMP_FOLDER = 'tmp/'
RES_FOLDER = 'res/'
print('Folder "{}" will be used to save temporary dictionary and corpus.'.format(TEMP_FOLDER)) 
import warnings 
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim') 
from gensim import corpora 
import gensim 
import numpy as np 
import pandas as pd 
import csv 
import collections 
from smart_open import open
import random 
from pprint import pprint  # pretty-printer 
import json


SIZE_OF_CHUNK = 5000


#function definition 
def read_corpus(f): 
    count = 0;
    for jsonLine in f :
        
        line = json.JSONDecoder().decode(jsonLine)
        doc_id = line['id']
        
        if count == 0 : print(doc_id)
        
        for morphs in line['morphs'] :
            yield gensim.models.doc2vec.TaggedDocument(morphs, [doc_id])  
            
        count = count + 1
        if count >= SIZE_OF_CHUNK : break
        
            
#file open_learning data file 
resFolder = 'res/'
contentsSize = 7
contents = [ open(os.path.join(resFolder, 'contents/data.' + str(i)), 'r') for i in range(contentsSize) ]
# contents = [ os.path.join(resFolder, 'contents/data.' + str(i)) for i in range(contentsSize) ]
contentsLineSize = [ 100000, 100000, 100000, 100000, 100000, 100000, 42190 ]


with open(TEMP_FOLDER+'wordDic_all_word', 'r') as wdf :
    wordDic_all_word = json.JSONDecoder().decode(wdf.read())
with open(TEMP_FOLDER+'wordDic_all_docid', 'r') as wdf :
    wordDic_all_docid = json.JSONDecoder().decode(wdf.read())

# metadata = pd.read_json(RES_FOLDER + 'metadata.json', lines=True)
# docIdList = [ w.strip() for w in metadata['id'] ]

print('Make [ wordList, docIdList ] finished.')

#build a model 
model = gensim.models.doc2vec.Doc2Vec(
    alpha=0.025,
    min_alpha=0.025,
    window=5,
    iter=4,
    size=20,
    
) 
#build_corpus = [ gensim.models.doc2vec.TaggedDocument(wordList, [docId]) for docId in docIdList ]
build_corpus = [ gensim.models.doc2vec.TaggedDocument(w, [wordDic_all_word[w]]) for w in wordDic_all_word ]
build_corpus.extend( [ gensim.models.doc2vec.TaggedDocument(wordDic_all_docid[d], [d]) for d in wordDic_all_docid ] )


model.build_vocab( build_corpus )
%time model.train(build_corpus, total_examples=model.corpus_count, epochs=1)
model.save( TEMP_FOLDER+'model_test_1.model') 

print('Build vocab initialize finished.')

for i in range(contentsSize) :
    
    for j in range( int(( contentsLineSize[i] + SIZE_OF_CHUNK - 1 ) / SIZE_OF_CHUNK ) ) :
        # read corpus
        train_corpus = list(read_corpus(contents[i]))         
        
        #load the model
        #model = gensim.models.Doc2Vec.load( TEMP_FOLDER+'model_test_1.model' )  # you can continue training with the loaded model! 
            
        #model.build_vocab( train_corpus, update=True )
        %time model.train(train_corpus, total_examples=model.corpus_count, epochs=model.iter)
        
        #save a model and word, sentence vector
        model.save( TEMP_FOLDER+'model_test_1.model') 
        #model.save_word2vec_format( TEMP_FOLDER+'model_test_1_format', doctag_vec=True, word_vec=True, prefix='*dt_', fvocab=None, binary=False) 
        
        
    contents[i].close()

```
